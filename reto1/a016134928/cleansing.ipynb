```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/AaTecM/semestre_7mo/septimo/reto1/input/train.csv')
df.head()
```

```python
#Describe df by column
df.describe(include='all')
```

```python
#Describe type of columns in df
df.dtypes
```

```python
#Delete empty spaces at the begining and in the end.
#Convert uppercase of strings to avoid inconsistences
df_obj = df.select_dtypes(['object'])
df[df_obj.columns] = df_obj.apply(lambda x: x.str.strip())
df[df_obj.columns] = df_obj.apply(lambda x: x.str.upper())
```

```python
#Validate NaN's in df
df_dict = {}
print('total of rows: ' + str(len(df['PassengerId'])))
for i in df.columns:
  #print(i)
  print('NaN of ' + str(i) +': ' + str(df[i].isna().sum()))
```

```python
#Delete 'Cabin' column due to high incidences in NaN
df = df.drop(columns = ['Cabin'])
```

```python
#Drop NaN rows
#The df is big enough to drop 179 values without compromise a futher analisis 
df = df.dropna()
df
```
